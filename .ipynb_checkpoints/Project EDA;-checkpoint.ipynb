## 3. 🧹 Data Cleaning

### Why We Do This
Real-world data is never clean. Cleaning is **not busywork** — it directly affects the accuracy of every analysis downstream. We need to:

1. **Quantify missing data** — decide whether to drop, fill, or flag each missing column
2. **Fix dirty strings** — this dataset has duplicated text in some columns (e.g., `'ActionAction'`)
3. **Parse messy formats** — dates and durations are stored as plain text; we need them as numbers
4. **Standardize categorical values** — avoid counting the same thing as two different categories

> **Production Rule:** Never modify the original DataFrame in place without first understanding what you're changing and why.

## 3.1 Missing Value Audit

Calculate both the **count** and **percentage** of missing values for each column.

> ⚠️ **Why percentage matters more than count:**
>
> - 50 missing values out of 100 → **Critical (50%)**
> - 50 missing values out of 1,000,000 → **Negligible (0.005%)**

### Key Idea
Always prioritize **percentage of missing data** over raw counts when assessing data quality.

## 3.3 Fix Duplicated Text in String Columns

The dataset contains duplicated text patterns caused by scraping errors, such as:

- `ActionAction`
- `ShounenShounen`

### Objective

Clean these values by removing repeated text while preserving the original meaning.

### Approach

Use a **reusable cleaning function** with **regular expressions (regex)** to:

- Detect repeated patterns in strings  
- Replace them with a single, correct value  

### Why This Matters

- Improves **data consistency**
- Prevents errors in **analysis and modeling**
- Ensures cleaner **categorical features**

```python
import re
text = "ActionAction"
match = re.sub(r'\b(\w+)\1\b',r'\1',text)
```

```python
match
```

```python
df_clean.head()
```

```python
def fix_doubled(col):
  if not isinstance(col,str):
    return col
  return re.sub(r'\b(\w+)\1\b',r'\1',col)
for i in ['Genres','Demographic']:
  df_clean[i] = df_clean[i].apply(fix_doubled)
```

```python
df_clean.head()
```

## 3.4 Parse Duration → Minutes (Numeric)

The `Duration` column contains text-based values such as:

- `24 min. per ep.`
- `1 hr. 44 min.`

### Objective

Convert these values into a **single numeric column (in minutes)**.

### Why This Matters

- Enables **mathematical operations** (e.g., averages, comparisons)
- Improves **model compatibility** (ML models require numeric input)
- Standardizes inconsistent time formats

### Approach

- Extract hours and minutes from the text  
- Convert everything into **total minutes**  
- Store the result in a new numeric column (e.g., `Duration_Min`)

```python
def duration(dur):
  hours = re.search(r'\b(\d+)\s*hr\b',dur)
  minutes = re.search(r'\b(\d+)\s*min\b',dur)
  # print(int(hours.group(1))*60)
  # print(int(minutes.group(1)))
  total = 0
  if hours:
    total += int(hours.group(1))*60
  if minutes:
    total += int(minutes.group(1))
  return float(total) if total > 0 else np.nan
df_clean['Duration_min'] = df_clean['Duration'].apply(duration)
```

## 3.5 Parse Episodes → Integer

The `Episodes` column is stored as a **string (object type)** instead of a numeric format.


Convert this column into an **integer (numeric type)** for proper analysis.

### Approach

Use `pd.to_numeric()` with `errors='coerce'`:

- Converts valid numeric values to integers  
- Replaces invalid or unparseable values with `NaN` (instead of throwing errors)

### Why This Matters

- Ensures **correct data types** for analysis  
- Prevents runtime errors during transformations  
- Safely handles messy or inconsistent data

```python
df_clean.info()
```

```python
df_clean['Episodes'] = pd.to_numeric(df_clean['Episodes'],errors="coerce")
```

## 3.6 Parse Air Start Year

The `Aired` column contains date ranges or single dates, such as:

- `Sep 29, 2023 to Mar 22, 2024`
- `Jan 8, 2021`

### Objective

Extract the **start year** from these values into a separate column.

### Why This Matters

- Enables **trend analysis over years/decades**
- Simplifies **time-based grouping and visualization**
- Reduces complexity from full date strings

### Approach

- Identify the **starting date** in the string  
- Extract the **year** from that date  
- Store it in a new column (e.g., `Start_Year`)

```python
def extract_year(date):
  match = re.search(r'\b(\d{4})\b',date)
  return match.group(1)
df_clean['Start_Year'] = df_clean['Aired'].apply(extract_year)
```

### Approach

- Identify common rating patterns  
- Map them to **simplified labels** (e.g., shorter categories)

## 6. ⚙️ Feature Engineering

### Why We Do This
Feature engineering is the process of creating **new columns** from existing ones to make the data more useful for analysis (and later, modeling). Raw data rarely contains the exact information you need — you have to derive it.

This is one of the **highest-leverage skills** a data analyst has. A well-engineered feature can make a simple model outperform a complex one.

We'll create:
- `Score_Category` — label anime quality tier
- `Is_Long_Running` — flag for 50+ episode shows
- `Total_Runtime_Hours` — total viewing time per anime
- `Decade` — group by decade for historical trends
- `Genre_Count` — how many genres each anime belongs to
- `Popularity_Score_Ratio` — normalized popularity vs quality metric
- `Has_English_Title` — binary flag

##6.1 Score Category (Binning)
## pd.cut() converts continuous values into labeled buckets
## This is called 'discretization' — useful for grouping analysis

```python
score = [0,7.99,8.24,8.49,8.74,10]
score_labels = ['Good(0-7.99)','Great(8.0-8.24)','Excellent(8.25 - 8.49)',
                'Outstanding(8.5 - 8.74)','Masterpiece(9+)']
df_clean['Score_category'] = pd.cut(
    df['Score'],
    bins=score,
    labels=score_labels
    )
```

## 7. 📐 Correlation Analysis

### Why We Do This
Correlation measures the **strength and direction** of the linear relationship between two numeric variables. It's a fast way to screen for which features might predict a target variable (like Score).

- **+1.0** = perfect positive correlation
- **0.0** = no linear relationship
- **-1.0** = perfect inverse correlation

⚠️ **Important:** Correlation ≠ Causation. Always think critically about *why* two variables correlate.

```python
df_clean.info()
```

```python
numeric_cols = [
    'Score','Popularity','Rank','Members','Episodes','Duration_min',
    'Start_Year','Genre_count','Long_Running','Total_Runtime'
]
corr = df_clean[numeric_cols].corr()
sns.heatmap(corr,annot=True)
```

## 8. 🔍 Outlier Detection

### Why We Do This
Outliers are data points that deviate significantly from the rest. They can:
- **Corrupt** statistical summaries (mean, std are sensitive to outliers)
- **Represent genuine rare events** worth investigating (legendary anime, data errors)
- **Break ML models** if not handled properly

We use two methods:
1. **IQR Method** — any point below Q1 - 1.5×IQR or above Q3 + 1.5×IQR is an outlier
2. **Z-Score Method** — any point > 3 standard deviations from the mean is an outlier
