# Filtering a DataFrame

```python
#q.find all the final winners
ipl.head(2)
```

```python
mask = ipl['MatchNumber'] == 'Final'
new_df = ipl[mask]
new_df[['Season','WinningTeam']]
```

```python
ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
# how many super overs finishes have occured
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# toss winner is match winner in percentage
(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python
# movies with rating higher than 8 and votes>1000
movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)].shape[0]
```

```python
# Action movies with rating higher than 7.5
movies[(movies['imdb_rating'] > 7.5) & (movies['genres'].str.split('|').apply(lambda x:'Action' in x))].shape[0]
```
